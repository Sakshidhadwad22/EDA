# Handling missing values

```python
df["normalized-losses"].unique()
```

```python
df["horsepower"].unique()
```

```python
df["normalized-losses"].replace("?",np.nan,inplace=True)
df["horsepower"].replace("?",np.nan,inplace=True)
```

```python
df["normalized-losses"]=df["normalized-losses"].astype(float)
df["horsepower"]=df["horsepower"].astype(float)
```

```python
df["normalized-losses"].fillna(df["normalized-losses"].mean(),inplace=True)
df["horsepower"].fillna(df["horsepower"].mean(),inplace=True)
```

# skewness removal

```python
colname=df.select_dtypes(["int","float"]).columns
colname
```

```python
from scipy.stats import skew
```

```python
for col in df[colname]:
    print(col)
    print(skew(df[col]))
    
    plt.figure()
    sns.distplot(df[col])
    plt.show()
```

```python
df.corr().style.background_gradient()
```

## to remove skewness we need two things 1st in skew value which is close to 0 and 2d thing is correlation which is close to 1 and when both condition is not satisfied that time we remove skewness 

## there are 2 ways to remove skewness by taking 1) log  OR 2) square root

```python
df["normalized-losses"]=np.sqrt(df["normalized-losses"])
```

```python
skew(df["normalized-losses"])
```

# Handling categorical data (Encoding)

```python
catcol=df.select_dtypes(object).columns
catcol
```

### there are 3 main types of encoder are there 1) One hot encoder 2) Label encoder 3) Ordinal encoder

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
df[catcol]=oe.fit_transform(df[catcol])
```
